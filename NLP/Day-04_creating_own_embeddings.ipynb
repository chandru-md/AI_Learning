```python
vocabulary_size = 5000
```

```python
messages = x.copy()
```

```python
messages.reset_index(inplace=True)
```

```python
import nltk
import re
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
```

```python
from nltk import corpus
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
corpus = []
for i in range(0,len(messages)):
  review = re.sub('[^a-zA-Z]',' ',messages['title'][i])
  review = review.lower()
  review = review.split()
  review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
  review = " ".join(review)
  corpus.append(review)
```

```python
corpus
```

```python
corpus[100]
```

```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
one_hot = [one_hot(words,vocabulary_size) for words in corpus]
```

```python
one_hot
```

```python
one_hot[100]
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
sen_len = 20
embedded_docs = pad_sequences(one_hot,maxlen=sen_len,padding='pre')
print(embedded_docs)
```

```python
embedded_docs = pad_sequences(one_hot,maxlen=sen_len,padding='post')
print(embedded_docs)
```
