## piviot, groupby, crosstab

```python
data=pd.DataFrame({'Category':['a','a','b','c','c','d'],"color":['red','blue','red','blue','yellow','green']})
```

```python
data
```

```python
pd.crosstab(data['Category'],data['color'])
```

```python
data_1=pd.DataFrame({"Product":['a','b','c','a','c','c'],"Sales":[100,200,130,609,880,976]})
```

```python
data_1
```
