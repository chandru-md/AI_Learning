```python
class Software_Emp:

    job_type = "Software Engineer"


    
    def __init__(self,Name,Emp_id,Designation,Salary,Address): #instant_variable
        self.Name=Name
        self.Emp_id=Emp_id
        self.Designation=Designation
        self.__Salary=Salary  #Encapsulation Method
        self.Address=Address

    def show_Salary(self,passcode):  #Getter
        if passcode =="Chan@1008":
            print(f"Salary of {self.Name} is {self.__Salary}")
        else:
            print("Access Denied")
    def increment_Salary(self,passcode): #Setter
        add_Salary=20000    #  local_variable
        if passcode == "Code@100":
            self.__Salary=self.__Salary+add_Salary
        else:
            print("Get lost")
            
```

```python
class Software_Emp:

    job_type = "Software Engineer"


    
    def __init__(self,Name,Emp_id,Designation,Salary,Address,Age): #instant_variable
        self.Name=Name
        self.Emp_id=Emp_id
        self.Designation=Designation
        self.__Salary=Salary  #Encapsulation Method
        self.Address=Address
        self.Age=Age

    def show_Salary(self,passcode):  #Getter
        if passcode =="Chan@1008":
            print(f"Salary of {self.Name} is {self.__Salary}")
        else:
            print("Access Denied")
    def increment_Salary(self,passcode): #Setter
        add_Salary=20000    #  local_variable
        if passcode == "Code@100":
            self.__Salary=self.__Salary+add_Salary
        else:
            print("Get lost")


class Hardware_Emp(Software_Emp):   # Hardware_Emp is inherited the Software_Emp from previous class.

    job_type="Hardware"

    def __init__(self,Name,Emp_id,Designation,Salary,Address,Age,Insur):
        Software_Emp. __init__(self,Name,Emp_id,Designation,Salary,Address,Age)
        self.__Salary=Salary
        self.__Insur=Insur

    def increment_Salary(self,passcode): #Setter
        add_Salary=20000    #  local_variable
        if passcode == "Code@100":
            self.__Salary=self.__Salary+add_Salary
            self.__Insur=self.__Insur + 30000
        else:
            print("Get lost")


    def show_Insur(self,passcode):  #Getter
        if passcode =="Chan@1008":
            print(f"Insurance of {self.Name} is {self.__Insur}")
        else:
            print("Access Denied")

    
```
