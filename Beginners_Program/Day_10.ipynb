```python
plt.figure(figsize=(10,5))
plt.hist(x="mpg",data=mt_cars_df)
plt.xlabel("MPG")
plt.ylabel("Freq")
plt.title("MPG Distribution",size=15)
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.hist(x="disp",data=mt_cars_df)
plt.xlabel("disp")
plt.ylabel("Freq")
plt.title("disp Distribution",size=15)
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.hist(x="cyl",data=mt_cars_df)
plt.xlabel("cyl")
plt.ylabel("Freq")
plt.title("cyl Distribution",size=15)
plt.show()
```

```python
plt.boxplot(x='mpg',data=mt_cars_df)
plt.show()
```

## 2-univariate analyse discrete data --count plot 

```python
import seaborn as sns
```

```python
sns.countplot(x='cyl',data=mt_cars_df)
plt.show()
```

```python
plt.pie(x=mt_cars_df['cyl'].value_counts(),autopct='%1.0f%%',labels=[4,5,6],explode=[0.02,0,0.04])
plt.show()
```
