```python
class Student_info:
    def __init__(self,name,address,phone,age):
        self.name=name
        self.address=address
        self.phone=phone
        self.age=age
    def greet(self):
        print(f"Hello {self.name} welcomes you, you are number is {self.phone}, you are age is {self.age}")

    def increased_by_10(self):
        self.age= self.age + 10
        print(self.age)
        
        
```

```python
Student_1.greet()
```

```python
Student_1.age
```

```python
Student_1.increased_by_10()
```
