```python
def print_numbers(*args):
    for num in args:
        print(num)
```

```python
print_numbers(1,2,3,4,5)
```

```python
def print_details(**chandru):
    for key,value in chandru.items():
        print(f"{key}: {value}")
```

```python
print_details(name="Chandru", age=25, city="Chennai")
```

```python
print_numbers(1,2,3,4,5)
print_numbers(10,20,30)
```

```python
print_details(name="sweety", age=30, country="USA")
```

```python
def print_details(*md, **chandru):
    for val in md:
        print(f"P.S : {val}")
    for key,value in chandru.items():
        print(f"{key}: {value}")
```

```python
print_numbers(1,2,3,4,5,"chandru","sweety")
```

```python
print_details("Hello", "Welcome", first_name="sweety", age=30, country="USA")
```
