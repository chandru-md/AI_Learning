```python
#union  === combines everything


set1 = {1,2,3,4,5,6}
set2 = {4,5,6,7,8,9}

union_set = set1.union(set2)

print(union_set)
```

```python
# intersection === pull out the common items
```

```python
set1 = {1,2,3,4,5,6}
set2 = {4,5,6,7,8,9}

inter_set = set1.intersection(set2)

print(inter_set)
```

```python
set1.intersection_update(set2)
print(set1)
```

```python
students = {
    'student1':{'name':'Chandru','age':12,'loc':'IND'},
    'student2':{'name':'Sweety','age':8,'loc':'UAE'},
    'student3':{'name':'Omana','age':7,'loc':'USA'},
    'student4':{'name':'Kavitha','age':13,'loc':'IND'}
}
```

```python
students.values()
```

```python
students.keys()
```

```python
for student_id,student_info in students.items():
    print(f"{student_id}:{student_info}")
    for key,value in student_info.items():
        print(f"{key}:{value}")
```
